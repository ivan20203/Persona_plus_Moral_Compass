```python
import os
import requests

# Talk to your Ollama server
os.environ["OLLAMA_HOST"] = "localhost:11434"
OLLAMA_HOST = os.environ["OLLAMA_HOST"]

# List all available models (equivalent to ollama.list())
response = requests.get(f"http://{OLLAMA_HOST}/api/tags")
models = response.json().get("models", [])

# Print each model name
for model in models:
    print(model.get("name"))
```
